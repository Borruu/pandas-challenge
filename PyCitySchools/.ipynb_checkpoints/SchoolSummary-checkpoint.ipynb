```python
import pandas as pd

# File to Load (Remember to Change These)
path1 = "Resources/schools_complete.csv"
path2 = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
schools = pd.read_csv(path1)
students = pd.read_csv(path2)

# Combine the data into a single dataset.  
df = pd.merge(students, schools, how="left", on=["school_name", "school_name"])
df.head()
```

```python
df.columns
```

```python
# Column: School name
# School Type
# Total School Budget

red_df = df.drop_duplicates(subset=["school_name"],ignore_index=True)
red_df = red_df[["school_name","School ID","type","size","budget"]]
red_df
schools_list = df["school_name"].value_counts()
#schools_list
#School Name[]
```

```python

perSt_list = red_df["budget"]/red_df["size"]

red_df["Budget per Student"]=perSt_list
red_df
```

```python
m_score_df =df.groupby(["school_name"])['maths_score'].sum()
#m_score_df.set_index("school_name")
m_score_df
```

```python
# Average maths score
red_df = red_df.sort_values(["school_name"])
red_df.reset_index()
mdf= pd.merge(red_df,m_score_df,on="school_name")

mdf["Average Maths Score"] = mdf["maths_score"]/mdf["size"]

mdf
#del red_df["index"]
```

```python
# Average reading score
r_score_df =df.groupby(["school_name"])['reading_score'].sum()
mdf = pd.merge(mdf,r_score_df, on="school_name")
mdf

```

```python
# Average reading score
mdf["Average Reading Score"] = mdf["reading_score"]/mdf["size"]
mdf
```
