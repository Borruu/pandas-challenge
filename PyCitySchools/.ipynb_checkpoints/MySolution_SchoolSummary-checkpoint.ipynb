```python
import pandas as pd

# File to Load (Remember to Change These)
path1 = "Resources/schools_complete.csv"
path2 = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
schools = pd.read_csv(path1)
students = pd.read_csv(path2)

# Combine the data into a single dataset.  
df = pd.merge(students, schools, how="left", on=["school_name", "school_name"])
df.head()
```

```python
df.columns
```

```python
# Column: School name
# School Type
# Total School Budget

red_df = df.drop_duplicates(subset=["school_name"],ignore_index=True)
red_df = red_df[["school_name","School ID","type","size","budget"]]
red_df
schools_list = df["school_name"].value_counts()
#schools_list
#School Name[]
```

```python

perSt_list = red_df["budget"]/red_df["size"]

red_df["Budget per Student"]=perSt_list
red_df
```

```python
m_score_df =df.groupby(["school_name"])['maths_score'].sum()
#m_score_df.set_index("school_name")
m_score_df
```

```python
# Average maths score
red_df = red_df.sort_values(["school_name"])
red_df.reset_index()
mdf= pd.merge(red_df,m_score_df,on="school_name")

mdf["Average Maths Score"] = mdf["maths_score"]/mdf["size"]

mdf
#del red_df["index"]
```

```python
# Average reading score
r_score_df =df.groupby(["school_name"])['reading_score'].sum()
mdf = pd.merge(mdf,r_score_df, on="school_name")
mdf

```

```python
# Average reading score
mdf["Average Reading Score"] = mdf["reading_score"]/mdf["size"]
mdf
```

```python
# % passing maths (the percentage of students who passed maths)
pm_df = df.loc[df["maths_score"]>=50,:]
# pm_c = pm_df.groupby(["school_name"])("student_name").count()
# pm_c

pm_c = pm_df.groupby(["school_name"])["student_name"].count()
pm_c
mdf = pd.merge(mdf,pm_c, on="school_name")
mdf.rename({"student_name":"# of Students passing Maths"})
mdf
# % passing reading (the percentage of students who passed reading)

# % overall passing (the percentage of students who passed maths AND reading)
```

```python
mdf = mdf.rename(columns = {'student_name':'# of Students passing Maths'})
mdf
```

```python
mdf["% of Students passing Maths"]=mdf["# of Students passing Maths"]/mdf["size"]*100
mdf.head()
```

```python
pr_df = df.loc[df["reading_score"]>=50,:]
# pm_c = pm_df.groupby(["school_name"])("student_name").count()
# pm_c

pr_c = pr_df.groupby(["school_name"])["student_name"].count()
pr_c
mdf = pd.merge(mdf,pr_c, on="school_name")
```

```python
mdf=mdf.rename(columns ={"student_name":"# of Students passing Reading"})
mdf.head()
```

```python
mdf["% of Students passing Reading"]=mdf["# of Students passing Reading"]/mdf["size"]*100
mdf.head()
```

```python
pb_df = df.loc[(df["maths_score"]>=50)&(df["reading_score"]>=50),:]
# pm_c = pm_df.groupby(["school_name"])("student_name").count()
# pm_c

pb_c = pb_df.groupby(["school_name"])["student_name"].count()

mdf = pd.merge(mdf,pb_c, on="school_name")
mdf.head()
```

```python
mdf = mdf.rename(columns={"student_name":"# of Students Passing"})
mdf.head()
```

```python
mdf["% of Students Passing"]=mdf["# of Students Passing"]/mdf["size"]*100
mdf.head()
```

```python
per_school_summary = mdf[["school_name","type","size","budget","Budget per Student","Average Maths Score","Average Reading Score","% of Students passing Maths","% of Students passing Reading","% of Students Passing"]]
per_school_summary.head()
```

```python
# Sort the schools by % Overall Passing in descending order (2 points)
# Save the results to a DataFrame called top_schools (2 points)
# Display the first 5 rows (1 point)
top_schools = per_school_summary.sort_values('% of Students Passing',ascending=False)
top_schools = top_schools.reset_index()
top_schools = top_schools.drop(columns=['index'])
                          
top_schools.head()
```

```python
# Sort the schools by % Overall Passing in ascending order (2 points)
# Save the results to a DataFrame called bottom_schools (2 points)
# Display the first 5 rows (1 point)
bottom_schools = per_school_summary.sort_values('% of Students Passing',ascending=True)
bottom_schools = bottom_schools.reset_index()
bottom_schools = bottom_schools.drop(columns=["index"])
bottom_schools.head()
```

```python
# Create a table that lists the average maths score for students of each year level (9, 10, 11, 12) at each school.
df.columns
df['year'].value_counts()
# Create a pandas series for each year. Hint: use a conditional statement.
y09 = df.loc[df['year']==9,:]
y10 = df.loc[df['year']==10,:]
y11 = df.loc[df['year']==11,:]
y12 = df.loc[df['year']==12,:]

# Group each series by school
y09gr = y09.groupby(["school_name"])["maths_score"].mean()
y10gr = y10.groupby(["school_name"])["maths_score"].mean()
y11gr = y11.groupby(["school_name"])["maths_score"].mean()
y12gr = y12.groupby(["school_name"])["maths_score"].mean()

# Combine the series into a dataframe
maths_scores_by_year = pd.DataFrame({"Year 9":y09gr,"Year 10":y10gr,"Year 11":y11gr,"Year 12":y12gr})
maths_scores_by_year
```

```python
# Optional: give the displayed data cleaner formatting
maths_scores_by_year['Year 9']=maths_scores_by_year['Year 9'].map("{:.2f}".format)
maths_scores_by_year['Year 10']=maths_scores_by_year['Year 10'].map("{:.2f}".format)
maths_scores_by_year['Year 11']=maths_scores_by_year['Year 11'].map("{:.2f}".format)
maths_scores_by_year['Year 12']=maths_scores_by_year['Year 12'].map("{:.2f}".format)
```

```python
# Create a table that breaks down school performances based on average Spending Ranges (Per Student). \
# Use 4 reasonable bins to group school spending. Include in the table each of the following:
    # Average Maths Score
    # Average Reading Score
    # % Passing Maths
    # % Passing Reading
    # Overall Passing Rate (Average of the above two)
per_school_summary
```

```python
bins = [0, 585, 630, 645, 680]
```

```python
label_list =["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
per_school_summary["Spending Ranges (Per Student)"] = pd.cut(per_school_summary['Budget per Student'],bins,labels=label_list,include_lowest=True)
per_school_summary
```

```python
spending_gr = per_school_summary.groupby("Spending Ranges (Per Student)").mean()
spending_gr
```

```python
spending_summary = spending_gr[["Average Maths Score","Average Reading Score","% of Students passing Maths",\
                           "% of Students passing Reading","% of Students Passing"]]
spending_summary
```

```python

size_labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
size_bins = [0,1000,2000,5000]

per_school_summary["School size"] = pd.cut(per_school_summary["size"],size_bins,labels=size_labels,include_lowest=False)
per_school_summary
```

```python
size_summary = per_school_summary.groupby("School size").mean()
size_summary
```

```python
size_summary = size_summary[["Average Maths Score","Average Reading Score","% of Students passing Maths",\
                           "% of Students passing Reading","% of Students Passing"]]
size_summary
```

```python
type_summary = per_school_summary.groupby("type").mean()
type_summary = type_summary[["Average Maths Score","Average Reading Score","% of Students passing Maths",\
                           "% of Students passing Reading","% of Students Passing"]]
type_summary

# per_school_summary["School size"] = pd.cut(per_school_summary["size"],size_bins,labels=size_labels,include_lowest=False)
# per_school_summary
```
